Vamos comear chamando as nossas funes auxiliares:

````python
# Funes auxiliares
def print_is_duplicated(df):
    """
    Funo que verifica se os dados esto duplicados. Se sim,
    mostra a quantidade de dados duplicados.
    """
    confirmation = df.duplicated().values.any()
    print(f"Existem valores duplicados? {confirmation}")
    if confirmation == True:
        print(f"Existem no total {df.duplicated().sum()} linhas duplicadas.")
    else:
        print('No existem linhas duplicadas.')


def print_lin_col(df):
  """
  Funo que imprime a quantidade de linhas e colunas do dataframe
  """
  print("O dataset possui:")
  print(f"{df.shape[0]} linhas")
  print(f"{df.shape[1]} colunas")
````

# Plotando grfico de pizza

Este grfico no  ideal quando temos muitos valores, pois isso compromete a visualizao dos dados. 
No entanto, para fins ilustrativos, escolhemos plotar o grfico de pizza em relao ao nvel de conhecimento em Python, no qual possuimos apenas 4 valores nicos.

Primeiramente, vamos alterar o nvel de conhecimento 'Nulo :D' para 'Sem experincia':

````python
mulheres['nivel_python'] = mulheres['nivel_python'].replace({'Nulo :D': 'Sem experincia'})
````

Em seguida, vamos criar uma tabela de contingncia, no qual podemos observar tanto a frequncia absoluta como a relativa:

````python
# Frequncia absoluta
nivel = mulheres['nivel_python'].value_counts(sort= True, ascending= False)

# Frequncia relativa
nivel_pct = (mulheres['nivel_python'].value_counts(sort= True, ascending= False, normalize= True) * 100)

# Cria o dataframe
lvl_python = pd.DataFrame([nivel, nivel_pct]).T.reset_index()
lvl_python.columns = ['nivel_python', 'frequencia', 'frequencia_pct']
lvl_python
````

Podemos definir a paleta de cores manualmente. Neste caso, usamos a tabela de cores em cdigo decimal.

````python
# Mudando as cores manualmente
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
````

Por fim, o nosso grfico:

````python
fig, ax = plt.subplots(figsize=(10, 8))
ax.pie(lvl_python.frequencia_pct,
       labels= lvl_python.nivel_python,
       autopct='%1.2f%%', startangle=180,
      textprops={'fontsize': 14},
      colors=colors)

plt.title("Nvel de conhecimento em Python", fontsize= 18)

plt.show()
````

# Exportando grficos

Podemos salvar os grficos como uma imagem .png ou .jpeg atravs do argumento ```plt.savefig('Nome_Imagem.png')```. O argumento ```dpi= ``` pode ajudar a melhorar a resoluo da imagem

<font color= 'red'> Ateno:</font> Este comando deve estar junto com o bloco de cdigo do grfico.

````python
fig, ax = plt.subplots(figsize=(10, 8))
ax.pie(lvl_python.frequencia_pct, labels= lvl_python.nivel_python,
       autopct='%1.2f%%', startangle=180,
      textprops={'fontsize': 14},
      colors=colors)

plt.title("Nvel de conhecimento em Python", fontsize= 18)

plt.show()

plt.savefig('pizza.png', dpi=300)
````
