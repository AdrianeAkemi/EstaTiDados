Esta primera etapa é igualzinha ao que aprendemos na aula, então não precisamos focar tanto nesta parte de limpeza dos dados.

Vamos começar chamando as nossas funções auxiliares:

````python
# Funções auxiliares
def print_is_duplicated(df):
    """
    Função que verifica se os dados estão duplicados. Se sim,
    mostra a quantidade de dados duplicados.
    """
    confirmation = df.duplicated().values.any()
    print(f"Existem valores duplicados? {confirmation}")
    if confirmation == True:
        print(f"Existem no total {df.duplicated().sum()} linhas duplicadas.")
    else:
        print('Não existem linhas duplicadas.')


def print_lin_col(df):
  """
  Função que imprime a quantidade de linhas e colunas do dataframe
  """
  print("O dataset possui:")
  print(f"{df.shape[0]} linhas")
  print(f"{df.shape[1]} colunas")
````

# Plotando gráfico de pizza

Este gráfico não é ideal quando temos muitos valores, pois isso compromete a visualização dos dados. 
No entanto, para fins ilustrativos, escolhemos plotar o gráfico de pizza em relação ao nível de conhecimento em Python, no qual possuimos apenas 4 valores únicos.

Primeiramente, vamos alterar o nível de conhecimento 'Nulo :D' para 'Sem experiência':

````python
mulheres['nivel_python'] = mulheres['nivel_python'].replace({'Nulo :D': 'Sem experiência'})
````

Em seguida, vamos criar uma tabela de contingência, no qual podemos observar tanto a frequência absoluta como a relativa:

````python
# Frequência absoluta
nivel = mulheres['nivel_python'].value_counts(sort= True, ascending= False)

# Frequência relativa
nivel_pct = (mulheres['nivel_python'].value_counts(sort= True, ascending= False, normalize= True) * 100)

# Cria o dataframe
lvl_python = pd.DataFrame([nivel, nivel_pct]).T.reset_index()
lvl_python.columns = ['nivel_python', 'frequencia', 'frequencia_pct']
lvl_python
````

Podemos definir a paleta de cores manualmente. Neste caso, usamos a tabela de cores em código decimal.

````python
# Mudando as cores manualmente
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
````

Por fim, o nosso gráfico:

````python
fig, ax = plt.subplots(figsize=(10, 8))
ax.pie(lvl_python.frequencia_pct,
       labels= lvl_python.nivel_python,
       autopct='%1.2f%%', startangle=180,
      textprops={'fontsize': 14},
      colors=colors)

plt.title("Nível de conhecimento em Python", fontsize= 18)

plt.show()
````

# Exportando gráficos

Podemos salvar os gráficos como uma imagem .png ou .jpeg através do argumento ```plt.savefig('Nome_Imagem.png')```. O argumento ```dpi= ``` pode ajudar a melhorar a resolução da imagem

<font color= 'red'> Atenção:</font> Este comando deve estar junto com o bloco de código do gráfico.

````python
fig, ax = plt.subplots(figsize=(10, 8))
ax.pie(lvl_python.frequencia_pct, labels= lvl_python.nivel_python,
       autopct='%1.2f%%', startangle=180,
      textprops={'fontsize': 14},
      colors=colors)

plt.title("Nível de conhecimento em Python", fontsize= 18)

plt.show()

plt.savefig('pizza.png', dpi=300)
````

Também é possível salvar em .pdf, que garante uma boa resolução:
